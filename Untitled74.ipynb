```python
import sys, os, numpy as np
sys.path.append("src")

import shap, torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from xtc.config import cfg

# Load trained checkpoint (local)
ckpt_path = "runs/latest"
tokenizer = AutoTokenizer.from_pretrained(ckpt_path, local_files_only=True)
model = AutoModelForSequenceClassification.from_pretrained(ckpt_path, local_files_only=True)
model.eval()

text = "NASA launches a new satellite into orbit"

def f(texts):
    # Normalize SHAP's inputs to List[str]
    if isinstance(texts, np.ndarray):
        texts = texts.astype(str).tolist()
    elif isinstance(texts, tuple):
        texts = list(texts)
    elif not isinstance(texts, list):
        texts = [str(texts)]
    else:
        texts = [str(t) for t in texts]

    toks = tokenizer(
        texts,
        padding=True,
        truncation=True,
        max_length=cfg.max_length,
        return_tensors="pt"
    )
    with torch.no_grad():
        logits = model(**toks).logits
    return logits.softmax(-1).cpu().numpy()

explainer = shap.Explainer(f, shap.maskers.Text(tokenizer))
sv = explainer([text])  # or: sv = explainer(np.array([text], dtype=object))
shap.plots.text(sv[0])  # inline visualization
```
